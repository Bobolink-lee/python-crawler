```python
import requests
import os
from bs4 import BeautifulSoup as bs
```

```python
# 创建目录
os.makedirs("pic/", exist_ok=True)
```

```python
# 下载图片
def download(file_path, picture_url, headers):
    img = requests.get(picture_url, headers=headers)
    with open(file_path, "wb") as f:
        f.write(img.content)
    # file = open(file_path, "wb")
    # file.write(img.content)
    # file.close()
```

```python
# 遍历pic_div
def travels(pic_div, headers):
    for i, div in enumerate(pic_div):
        filename = div.a.img["alt"]
        picture_url = div.a.img["src"]
        file_path = 'pic/' + filename + '.jpg'
        if os.path.exists(file_path):
            continue
        download(file_path=file_path, picture_url=picture_url, headers=headers)
        print(f'the {i+1}st file, {filename} already download')
```

```python
def crawler(url, headers):
    response = requests.get(url, headers=headers)
    html = response.text
    soup = bs(html)
    pic_div = soup.find_all(name="div", attrs={"class": "pic"})
    travels(pic_div=pic_div, headers=headers)
```

```python
def main():
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36 Edg/117.0.2045.36"
    }
    for i in range(0, 250, 25):
        url = f"https://movie.douban.com/top250?start={i}&filter="
        crawler(url=url, headers=headers)
```

```python
if __name__ == "__main__":
    main()
```
